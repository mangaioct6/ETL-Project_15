### Creating tables

```python
# To create tables under asteroids database, we have to make connection with our database
connection = psycopg2.connect(
             host = 'localhost',
         database = 'asteroids',
            user  = 'postgres',
         password = 'admin'
)

connection.autocommit = True
cursor = connection.cursor()


cursor.execute('DROP SEQUENCE IF EXISTS eventid')
cursor.execute('DROP TABLE IF EXISTS asteroid_details')
cursor.execute('DROP TABLE IF EXISTS asteroid_size')
cursor.execute('DROP TABLE IF EXISTS asteroid_speed')
cursor.execute('DROP TABLE IF EXISTS hazardous_asteroid')

#Same asteroids can come near earth multiple times. so creating an sequence for the events
cursor.execute("CREATE SEQUENCE eventid \
                            INCREMENT 1 \
                            MINVALUE 1 \
                            START 1")

cursor.execute("CREATE TABLE asteroid_details (event_id INT  PRIMARY KEY,\
                                               neo_reference_id INT,\
                                               asteroid_name varchar(255),\
                                               nasa_url varchar(255),\
                                               close_approach_date timestamp not null);")

cursor.execute("CREATE TABLE asteroid_size (neo_reference_id INT,\
                                            est_diameter_min_km numeric,\
                                            est_diameter_max_km numeric,\
                                            absolute_magnitude numeric,\
                                            close_approach_date_time timestamp not null);")

cursor.execute("CREATE TABLE asteroid_speed(neo_reference_id INT,\
                                            relative_velocity_km_sec numeric,\
                                            relative_velocity_km_hour numeric,\
                                            miss_distance_AU numeric);")

cursor.execute("CREATE TABLE hazardous_asteroid(neo_reference_id INT ,\
                                                close_approach_date_time timestamp not null,\
                                                hazardous_asteroid boolean);")

#closing connection
cursor.close()
connection.close()
```

### Inserting datas into tables

```python
connection = psycopg2.connect(
             host = 'localhost',
         database = 'asteroids',
            user  = 'postgres',
         password = 'admin'
)
connection.autocommit = True
cursor = connection.cursor()
        
asteroid_details_insert = ("INSERT INTO asteroid_details (event_id,\
                                                          neo_reference_id,\
                                                          asteroid_name,\
                                                          nasa_url,\
                                                          close_approach_date) \
                                                          VALUES(nextval('eventid'),%s,%s,%s,%s);")
asteroid_size_insert = ("INSERT INTO asteroid_size (neo_reference_id, \
                                                    est_diameter_min_km,\
                                                    est_diameter_max_km,\
                                                    absolute_magnitude,\
                                                    close_approach_date_time) \
                                                    VALUES (%s,%s,%s,%s,%s);")
asteroid_speed_insert = ("INSERT INTO asteroid_speed (neo_reference_id,\
                                                      relative_velocity_km_sec,\
                                                      relative_velocity_km_hour,miss_distance_AU)\
                                                      VALUES(%s,%s,%s,%s);")
hazardous_insert = ("INSERT INTO hazardous_asteroid (neo_reference_id, \
                                                     close_approach_date_time, \
                                                     hazardous_asteroid) \
                                                     VALUES (%s,%s,%s);")

#creating a function to insert datas into tables

def insert(insert, df):
    for i,row in df.iterrows():
        cursor.execute(insert,list(row))
        
#calling insert function
insert(asteroid_details_insert,asteroid_details_df)
insert(asteroid_size_insert,asteroid_size_df)
insert(asteroid_speed_insert,asteroid_speed_df)
insert(hazardous_insert,hazardous_asteroid_df )

cursor.close()
connection.close()
```

## Queries:

### 1. Show the total number of asteroids came closer to Earth from 2021-06-01 to 2021-12-31

* To find the total number of asteroids came near to Earth during calculated period, I am counting the neo reference id from asteroid_details tables and name as asteroid_count

```python
%sql select count(neo_reference_id) as asteroid_count from asteroid_details
```

### 2. Show how many asteroids came nearby Earth more than once during calculated period

* There is a chance, the same asteroid can pass by Earth more than once. To count unique neo_reference_id, using distinct keyword. Writing subquery in where clause with the condition of having neo_reference_id count > 1 from asteroid_details table.

```python
%sql select count(distinct neo_reference_id) from asteroid_details \
     where neo_reference_id in (\
     select neo_reference_id from asteroid_details \
     group by neo_reference_id \
     having count(*)>1)
```

### 3. Write a Query to show how many asteroids are hazardous and not hazardous. 

* I have selected hazardous_asteroid column from hazardous_asteroid table. Filtered hazardous and non hazardous in where clause and grouped by hazardous_asteroid.

```python
%sql select distinct hazardous_asteroid,count(*) from hazardous_asteroid \
            where  hazardous_asteroid = True or hazardous_asteroid = False \
            group by hazardous_asteroid 
```

### 4. Show the hazardous asteroids name, magnitude,est_diameter_max_km and relative_velocity(km/s).

* I have joined asteroid_details and asteroid_size using neo_reference_id to print asteroid_name& magnitude. I did inner join asteroid_size and hazardous_asteroid using neo_reference_id to find hazardous asteroid. Joined hazardous_asteroid and asteroid_speed to print velocity. I have rounded estimated diameter and relative velocity and filtered the asteroids which is hazardous using where clause.

```python
%sql select distinct asd.asteroid_name, asi.absolute_magnitude, \
     round(asi.est_diameter_max_km,2) as max_diameter_km, \
     round(asp.relative_velocity_km_sec,2) as relative_velocity \
     from asteroid_details as asd \
     inner join asteroid_size as asi \
     on asd.neo_reference_id = asi.neo_reference_id \
     inner join hazardous_asteroid as hd \
     on asi.neo_reference_id = hd.neo_reference_id \
     inner join asteroid_speed as asp \
     on hd.neo_reference_id = asp.neo_reference_id \
     where hazardous_asteroid = True
```

### 5. Write a query that shows how many asteroids were found each month. Group them by month.

* I wanted use datepart in my query. So I created a question to print the month wise asteroids count. I have selected month part from close_approach_date and named as month from asteroid_details table. I grouped and ordered by month. 

```python
%sql select date_part('month',close_approach_date) as month,count(distinct neo_reference_id) \
     from asteroid_details \
     group by month \
     order by count desc
```

### 6. Write a query to find the largest asteroid came near Earth during the calculated period.

* I have filtered the asteroids which is having maximum estimated diameter greater than 5 using where clause. To print the asteroid name, velocity and date I joined asteroid_size and asteroid_details tables. And then I have joined asteroid_speed and asteroid_details using neo_reference_id. 


```python
%sql select distinct asi.neo_reference_id, round(asi.est_diameter_max_km,2) as diameter_km, \
     asd.asteroid_name, round(asp.relative_velocity_km_sec,2) as velocity_km_s, \
     asi.close_approach_date_time   from \
     asteroid_size as asi \
     inner join asteroid_details as asd \
     on asi.neo_reference_id = asd.neo_reference_id \
     inner join asteroid_speed as asp \
     on asd.neo_reference_id = asp.neo_reference_id \
     where est_diameter_max_km > 5 
     
```

### 7. How fast does asteroids pass by Earth, on average?

* We can calculate how fast an asteroid can pass earth by taking average of relative velocity. I have named average relative velocity as velocity.

```python
%sql select round(avg(relative_velocity_km_sec),2) as velocity_km_per_sec from asteroid_speed
```
